```python
def quicksort(a, p, q):  
    if p < q:  
        j = partition(a, p, q+1)   # partition step, returns index j
        quicksort(a, p, j-1)       # sort left partition
        quicksort(a, j+1, q)       # sort right partition



def partition(a, m, p):  
    global comparisons

    v = a[m]      
    i = m       
    j = p     

    while True:

        # -------- Move i to the right --------
        while True:
            comparisons += 1          #
            i += 1                   

            if a[i] >= v:            
                break

        # -------- Move j to the left --------
        while True:
            comparisons += 1          
            j -= 1                    

            if a[j] <= v:             
                break

        # -------- Swap or exit --------
        if i < j:
            interchange(a, i, j)      
        else:
            break                   

    # -------- Place pivot in correct position --------
    interchange(a, m, j)
    return j     

def interchange(a, i, j):
    a[i], a[j] = a[j], a[i]
```

```python
a = [38, 27, 43, 3, 9, 82, 10]
# Add sentinel âˆž
a.append(float("inf"))

comparisons = 0   

quicksort(a, 0, len(a)-1)  

a.pop()   # remove sentinel

print("Sorted =", a)
print("Comparisons =", comparisons)
```
