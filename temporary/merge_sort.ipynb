```python
def mergesort(low, high):
    if low < high:
        mid = (low + high) // 2
        
        mergesort(low, mid)
        mergesort(mid + 1, high)
        
        merge(low, mid, high)               # Combine step

def merge(low, mid, high):
    global comparisons
    
    h = low          # pointer for left half
    i = low          # pointer for b[] (merged array)
    j = mid + 1      # pointer for right half

    # Compare elements until one list is exhausted
    while (h <= mid) and (j <= high):
        comparisons += 1
        if a[h] <= a[j]:
            b[i] = a[h]
            h = h + 1
        else:
            b[i] = a[j]
            j = j + 1
        i = i + 1     # always increase i no matter what

    # Copy remaining elements
    if h > mid:          # left half finished → copy right half
        for k in range(j, high + 1):
            b[i] = a[k]
            i = i + 1
    else:                # right half finished → copy left half
        for k in range(h, mid + 1):
            b[i] = a[k]
            i = i + 1

    for k in range(low, high + 1):        # Copy back to original array
        a[k] = b[k]

a = [38, 27, 43, 3, 9, 82, 10]
b = [0] * len(a)
comparisons = 0
mergesort(0, len(a) - 1)
print(a) 
print("Total comparisons =", comparisons)
```
