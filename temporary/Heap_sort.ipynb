```python
def heapify(arr, n, i, comparisons):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n:
        comparisons += 1
        if arr[left] > arr[largest]:
            largest = left

    if right < n:
        comparisons += 1
        if arr[right] > arr[largest]:
            largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        arr, comparisons = heapify(arr, n, largest, comparisons)

    return arr, comparisons

def heap_sort(arr):
    n = len(arr)
    comparisons = 0

    for i in range(n // 2 - 1, -1, -1):
        arr, comparisons = heapify(arr, n, i, comparisons)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        arr, comparisons = heapify(arr, i, 0, comparisons)

    return arr, comparisons

arr = [9, 5, 1, 4, 3]
sorted_arr, comps = heap_sort(arr)
print("Original array:", [9, 5, 1, 4, 3])
print("Sorted array:", sorted_arr)
print("Number of comparisons:", comps)
```
