```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    print(start, end=" ")     # display the node

    for neighbour in graph[start]:
        if neighbour not in visited:
            dfs(graph, neighbour, visited)
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("DFS Traversal:", end=" ")
dfs(graph, 'A')
```
