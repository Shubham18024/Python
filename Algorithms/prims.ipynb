```python
import heapq

def prim(graph, start):
    visited = set()
    pq = [(0, start, -1)]
    mst = []
    total = 0

    while pq:
        w, u, parent = heapq.heappop(pq)
        if u in visited:
            continue
        visited.add(u)
        if parent != -1:
            mst.append((parent, u, w))
            total += w

        for v, wt in graph[u]:
            if v not in visited:
                heapq.heappush(pq, (wt, v, u))

    return mst, total


graph = {
    0: [(1, 4), (7, 8)],
    1: [(0, 4), (2, 8)],
    2: [(1, 8), (3, 7)],
    3: [(2, 7)],
    7: [(0, 8)]
}

mst, cost = prim(graph, 0)
print("MST:", mst)
print("Total Cost:", cost)
```
