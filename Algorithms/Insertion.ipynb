```python
def insertion_sort(A,n):
    global comparisons

    for j in range(1, n):
        key = A[j]          
        i = j - 1         
        while i >= 0 and A[i] > key:
            comparisons += 1  
            A[i+1] = A[i] 
            i = i - 1
        # count the final comparison that made the loop stop
        if i >= 0:
            comparisons += 1

        A[i+1] = key         
```

```python
A = [80, 20, 40, 90, 30, 60]
comparisons = 0

insertion_sort(A,len(A))

print("Sorted array =", A)
print("Total comparisons =", comparisons)
```
