```python
class Sll_Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Sll_construction:
    def __init__(self):
        self.head = None
        
    def insertAtBegin(self, data):
        new_node = Sll_Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node
            
    def removeFirstNode(self):
        if self.head is None:
            return "Empty List"
        else:
            removed_data = self.head.data
            self.head = self.head.next
            return removed_data
            
    def traverseSLL(self):
        if self.head is None:
            print("The list is empty.")
        temp = self.head
        while temp is not None:
            print(temp.data, end=" --> ")
            temp = temp.next
        print(None)
```

```python
def main():
    llist = Sll_construction()
    print("*** Linked List ***")
    while True:
        print("\n1. Insert At Beginning")
        print("2. Delete At Beginning")
        print("3. Show Entire List")
        print("4. Exit")
        
        choice = int(input("Enter your choice: "))
        if choice == 1:
            value = eval(input("Enter the value to be inserted: "))
            llist.insertAtBegin(value)
            print(f"{value} inserted at the beginning.")
        elif choice == 2:
            removed_value = llist.removeFirstNode()
            if removed_value == "Empty List":
                print("Cannot delete. The list is empty.")
            else:
                print(f"Value Deleted: {removed_value}")
        elif choice == 3:
            llist.traverseSLL()
            
        elif choice == 4:
            print("Exiting program. Goodbye! (◔‿◔) ")
            break
        else:
            print("Invalid Choice. Please try again.")
        
if __name__ == "__main__":
    main()
```
